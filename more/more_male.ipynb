```python
def plot_box_two(arr_1,arr_2,column_1,column_2):
    fig, ax = plt.subplots(figsize=(8,5))
    ax.boxplot([arr_1,arr_2])
    plt.xticks([1, 2], [column_1, column_2], rotation=10)
    plt.show()
```

```python
def plot_error_distribution(array):
    plt.subplots(figsize=(8,5))
    #Set X-Axis range
    
    plt.title('Model Error Distribution')
    plt.ylabel('No. of Predictions')
    plt.xlabel('Error (cm)')
    plt.hist(array)
    plt.show()
```

```python
waist_target_arr=join[waist].to_numpy()
waist_pred_arr=join[waist_pred].to_numpy()
waist_error=(np.subtract(waist_target_arr,waist_pred_arr))/10
plot_error_distribution(waist_error)
plot_box_two(waist_target_arr/10,waist_pred_arr/10,waist,waist_pred)
plot_box_two(waist_error,[],'Waist Error',"")
```
